```python
def eps_dec(eps_decay_rate, i):
    return np.exp(-eps_decay * i)
```

```python
scores = []
test_scores = []
for i in range(200):
    score = 0
    done = False
    observation = env.reset()
    while not done:
        #agent.Q_eval.optimizer.zero_grad()
        action = agent.choose_action(observation,eps_dec(0.005,i)*0.2)
        observation_, reward, done, info = env.step(action)
        score += reward
        agent.store_tranisitions(observation, action, reward, observation_, done)
        
        agent.learn()
        observation = observation_

    scores.append(score)
    #eps_history.append(epsilon)
    
    avg_score = np.mean(scores[-100:])
    #print('episode:', i, 'score: %.2f'%score,'avg_score %.2f'%avg_score)
    

    observation = env.reset()
    done = False
    test_score = 0
    while done != True:
        action = agent.choose_action(observation,0)
        observation_, reward, done, info = env.step(action)
        agent.store_tranisitions(observation, action, reward, observation_, done)
        agent.learn()
        observation = observation_
        test_score += reward
    test_scores.append(test_score)
    print('episode: %3d' % i, 'score: %5.2f'%score,
          'avg_score %5.2f'%avg_score, 'test_score %5.2f'% test_score, 'avg_test_score: %5.2f'%np.mean(test_scores[-100:]))
        #env.render()
        
env.close()

```

```python
#plt.plot(test_scores)
plt.figure(figsize=(20,10))
plt.plot(ta.trend.ema(pd.Series(test_scores),100),label='test_scores',linewidth=4)
plt.plot(ta.trend.ema(pd.Series(scores),100), label = 'train_scores',linewidth=4)
plt.xlabel('episodes',fontsize=20)
plt.ylabel('score',fontsize=20)
plt.legend(fontsize=40)
```
